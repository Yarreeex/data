### Analisis klaster kabupaten Kroasia berdasarkan kriteria demografis

Di antara metode dan teknik analisis multivariat, metode k-means telah dipilih 
sebagai yang paling tepat untuk pengelompokan kabupaten Kroasia. Selain 
keunggulan lainnya, metode k-means sangat efektif karena memberikan peluang 
jumlah cluster dapat ditentukan terlebih dahulu. Umumnya metode k-means 
digunakan untuk mengelompokkan partisi n objek menjadi k cluster dimana 
setiap objek termasuk dalam cluster dengan mean terdekat. Metode ini 
menghasilkan tepat k kluster yang berbeda dengan perbedaan sebesar mungkin. 
Jumlah cluster k terbaik yang mengarah ke pemisahan (jarak) terbesar tidak 
dikenal sebagai apriori dan harus dihitung dari data. Tujuan penggunaan metode 
k-means untuk pengelompokan adalah untuk meminimalkan total varian intra cluster atau fungsi kesalahan kuadrat.
Algoritma metode k-means berlangsung dalam beberapa langkah. Yang pertama adalah 
mengelompokkan data ke dalam k grup di mana k sudah ditentukan sebelumnya. Pada langkah 
kedua, algoritma secara acak memilih k titik sebagai pusat klaster. Langkah ketiga adalah 
menetapkan objek ke pusat cluster terdekat sesuai dengan fungsi jarak Euclidean. Langkah 
keempat yang melibatkan algoritma metode k-means terdiri dari menghitung centroid atau rata rata dari semua objek di setiap cluster. Terakhir, langkah kelima mengulangi langkah 2, 3, dan 4 
hingga poin yang sama diberikan ke setiap cluster dalam putaran yang berurutan.
Oleh karena itu metode k-means merupakan metode clustering yang relatif efisien. Namun, 
kami perlu menentukan jumlah cluster terlebih dahulu. Sayangnya, tidak ada metode 
teoretis global untuk menemukan jumlah cluster yang optimal. Pendekatan praktis adalah 
membandingkan hasil dari beberapa proses dengan k yang berbeda dan memilih yang 
terbaik berdasarkan kriteria yang telah ditentukan. Secara umum, k besar mungkin 
mengurangi kesalahan tetapi meningkatkan risiko overfitting. Keluaran metode k-means 
secara teratur meliputi analisis statistik varians F. Sementara statistik ini bersifat 
oportunistik (prosedur mencoba untuk membentuk kelompok yang berbeda), ukuran relatif 
dari statistik memberikan informasi tentang kontribusi pemisahan masing-masing variabel 
terhadap pemisahan kelompok.
